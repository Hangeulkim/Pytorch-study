```python
import os
import shutil

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
image_path = 'C:/Users/USER/Desktop/chg/val'

chg_path = 'C:/Users/USER/AE_transient_Gabor_transform_IMF1_2_3mm/git_upload/images/none'
createFolder(chg_path)

folds=os.listdir(image_path)
#shutil.rmtree('C:/Users/USER/AE_transient_Gabor_transform_IMF1_2_3mm/git_upload/images')
for fold in folds:
    if '3mm' not in fold and 'normal' not in fold:
        continue
    files = os.listdir(image_path+'/'+fold)
    img_path = chg_path
    if '300' in fold:
        img_path+="/val"
        createFolder(img_path)
    elif '400' in fold:
        img_path+="/test"
        createFolder(img_path)
    else:
        img_path+="/train"
        createFolder(img_path)
    img_path+="/"+fold[fold.find('_')+1:fold.rfind('_')]
    createFolder(img_path)
    for file in files:
        if 'png' not in file and 'PNG' not in file:
            continue
        shutil.copy(image_path+"/"+fold+"/"+file,img_path+"/"+file[:file.rfind('.')]+"_"+fold[:fold.find('_')]+"_"+fold[fold.rfind('_')+1:]+".png")

    
```

```python
import os
from PIL import Image

image_path = 'images/none'
cutted_path = 'images/'
way=""
img_list = os.listdir(image_path)
 
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)


def fold_chk(directory):
    if os.path.isdir(image_path+directory):
        next_fold = os.listdir(image_path+directory)
        for fold in next_fold:
            if os.path.isdir(image_path+directory+"/"+fold):
                fold_chk(directory+"/"+fold)
                
            else: 
                if 'png' not in fold[fold.rfind('.')+1:] and 'PNG' not in fold[fold.rfind('.')+1:]:
                    continue
                    
                print(image_path+directory+"/"+fold)
                
                createFolder(cutted_path+"/none-112_single"+directory)
                top_left = cutted_path+"/top_left-64_single"+directory
                createFolder(top_left)
                top_right = cutted_path+"/top_right-64_single"+directory
                createFolder(top_right)
                bottom_left = cutted_path+"/bottom_left-64_single"+directory
                createFolder(bottom_left)
                bottom_right = cutted_path+"/bottom_right-64_single"+directory
                createFolder(bottom_right)
                middle = cutted_path+"/middle-64_single"+directory
                createFolder(middle)

                img = Image.open('%s%s/%s'%(image_path, directory,fold))
                img = img.resize((112,112))
                img_sz = img.size
                img.save('%s/none-112_single/%s/%s'%(cutted_path, directory, fold))
                
                cutted_img = img.crop((0,0,64,64))
                cutted_img.save(top_left+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]-64,0,img_sz[0],64))
                cutted_img.save(top_right+'/%s'%(fold))
                
                cutted_img = img.crop((0,img_sz[1]-64,64,img_sz[1]))
                cutted_img.save(bottom_left+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]-64,img_sz[1]-64,img_sz[0],img_sz[1]))
                cutted_img.save(bottom_right+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]/2-32,img_sz[1]/2-32,img_sz[0]/2+32,img_sz[1]/2+32))
                cutted_img.save(middle+'/%s'%(fold))
                
                createFolder(cutted_path+"/none-64_single"+directory)
                top_left = cutted_path+"/top_left-32_single"+directory
                createFolder(top_left)
                top_right = cutted_path+"/top_right-32_single"+directory
                createFolder(top_right)
                bottom_left = cutted_path+"/bottom_left-32_single"+directory
                createFolder(bottom_left)
                bottom_right = cutted_path+"/bottom_right-32_single"+directory
                createFolder(bottom_right)
                middle = cutted_path+"/middle-32_single"+directory
                createFolder(middle)

                img = Image.open('%s%s/%s'%(image_path, directory,fold))
                img = img.resize((64,64))
                img_sz = img.size
                img.save('%s/none-64_single/%s/%s'%(cutted_path, directory, fold))
                
                cutted_img = img.crop((0,0,32,32))
                cutted_img.save(top_left+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]-32,0,img_sz[0],32))
                cutted_img.save(top_right+'/%s'%(fold))
                
                cutted_img = img.crop((0,img_sz[1]-32,32,img_sz[1]))
                cutted_img.save(bottom_left+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]-32,img_sz[1]-32,img_sz[0],img_sz[1]))
                cutted_img.save(bottom_right+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]/2-16,img_sz[1]/2-16,img_sz[0]/2+16,img_sz[1]/2+16))
                cutted_img.save(middle+'/%s'%(fold))
                
                createFolder(cutted_path+"/none-32_single"+directory)
                top_left = cutted_path+"/top_left-16_single"+directory
                createFolder(top_left)
                top_right = cutted_path+"/top_right-16_single"+directory
                createFolder(top_right)
                bottom_left = cutted_path+"/bottom_left-16_single"+directory
                createFolder(bottom_left)
                bottom_right = cutted_path+"/bottom_right-16_single"+directory
                createFolder(bottom_right)
                middle = cutted_path+"/middle-16_single"+directory
                createFolder(middle)

                img = Image.open('%s%s/%s'%(image_path, directory,fold))
                img = img.resize((32,32))
                img_sz = img.size
                img.save('%s/none-32_single/%s/%s'%(cutted_path, directory, fold))
                
                cutted_img = img.crop((0,0,16,16))
                cutted_img.save(top_left+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]-16,0,img_sz[0],16))
                cutted_img.save(top_right+'/%s'%(fold))
                
                cutted_img = img.crop((0,img_sz[1]-16,16,img_sz[1]))
                cutted_img.save(bottom_left+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]-16,img_sz[1]-16,img_sz[0],img_sz[1]))
                cutted_img.save(bottom_right+'/%s'%(fold))
                
                cutted_img = img.crop((img_sz[0]/2-8,img_sz[1]/2-8,img_sz[0]/2+8,img_sz[1]/2+8))
                cutted_img.save(middle+'/%s'%(fold))
                
                createFolder(cutted_path+"/none-16_single"+directory)
                img = Image.open('%s%s/%s'%(image_path, directory,fold))
                img = img.resize((16,16))
                img_sz = img.size
                img.save('%s/none-16_single/%s/%s'%(cutted_path, directory, fold))
                
                
        return
    
    
for name in img_list :
    way="/"+name
    fold_chk(way)

shutil.rmtree(image_path)
```
